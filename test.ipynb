```python
import wbgapi as wb
import pandas as pd
import matplotlib.pyplot as plt
```

```python
countries = ['CHN', 'JPN']
indicators = {
    'SP.POP.65UP.TO.ZS': 'Aging_Rate',       # 65岁以上人口占比
    'GB.XPD.RSDV.GD.ZS': 'RD_Intensity',    # 研发支出占GDP比重
}

df = wb.data.DataFrame(indicators.keys(), countries, mrv=15).T
df.index = df.index.str.replace('YR', '').astype(int)

# 绘制双轴图，你可以看到中国R&D曲线的陡峭上升，对比日本的平稳。
# 同时也能看到中国老龄化曲线正在以比日本当年更快的斜率上升。
df.unstack().plot(figsize=(10, 6), title="China vs Japan: Tech vs Aging")
plt.show()
```
